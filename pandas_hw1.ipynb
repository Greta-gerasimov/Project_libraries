```python
import pandas as pd 
import numpy as np
```

```python
authors = pd.DataFrame({'authors_id': [1,2,3],
                        'authors_names': ['Тургенев', 'Чехов','Островский']},columns=['authors_id','authors_names'])
```

```python
authors
```

```python
books = pd.DataFrame({'authors_id': [1, 1 , 1, 2, 2 , 3, 3],
                      'book_title': ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],
                      'price' : [450, 300, 350, 500, 450, 370, 290]}, columns = ['authors_id','book_title', 'price'] )
```

```python
books
```

```python
authors_price = pd.merge(authors, books,on = 'authors_id', how = 'outer')
authors_price
```

```python
top5 = authors_price.nlargest(5,'price')
top5
```

```python
max_price = authors_price.groupby('authors_names').agg({"price": "max"})
max_price
```

```python
min_price = authors_price.groupby('authors_names').agg({"price":"min"})
min_price
```

```python
mean_price = authors_price.groupby('authors_names').agg({"price":"mean"})
mean_price
```

```python
authors_stat = pd.merge(max_price, min_price,  on = "authors_names", how = "outer")
authors_stat
```

```python
authors_stat = pd.merge(authors_stat, mean_price, on = "authors_names", how = "outer")
authors_stat
```

```python
authors_stat = authors_stat.rename(columns={'price_x':'max_price', 'price_y':'min_price','price':'mean_price'})
authors_stat
```

```python
authors_price['cover']= ('твердая' ,'мягкая','мягкая','твердая','твердая','мягкая','мягкая')
authors_price
```

```python
books_info = pd.pivot_table(authors_price, values='price', index=['authors_names'],  columns=['cover'], aggfunc=np.sum)
books_info
```

```python
books_info['мягкая'].fillna(0, inplace=True)
books_info
```

```python
books_info['твердая'].fillna(0, inplace = True)
books_info
```

```python
import pickle
book_info_pkl = pickle.dumps(books_info)
print(f'This is my pickled file:\n {book_info_pkl}')
```

```python
book_info_2 = pickle.loads(book_info_pkl)
print(f'This is my unpickled file:\n {book_info_2}')
```
